```python
import os
```

```python
path = 'adv_classes'
train_file='train_list.txt'
```

```python
path = 'adv_classes'
test_file='testing_list.txt'
test_file_new='new_testing_list.txt'
```

```python
data = list()
for root, dir, files in os.walk(path):
    for file in files:
        f_path = os.path.join(root, file)
        data.append(f_path)    
```

```python
# reading test files
with open(os.path.join(path,test_file)) as f:
    test_lines = f.readlines()
```

```python
# adding path to test files
test_lines_new=[]
for line in test_lines:
    test_lines_new.append(path+'/'+line)    
```

```python
test_lines_new
```

```python
# writing train file
with open(os.path.join(path,train_file), 'w') as f:
    for line in data:
        f.write(line)
        f.write('\n')
```

```python
# writing new test file
with open(os.path.join(path,test_file_new), 'w') as f:
    for line in test_lines_new:
        f.write(line)
```
